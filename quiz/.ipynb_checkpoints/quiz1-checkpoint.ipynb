# Q2


```python
import numpy as np
a=np.zeros((20,20))
b=np.ones((20,20))
def l2_dist(a, b):
     result = ((a - b) * (a - b)).sum()
     result = result ** 0.5
     return result
```

```python
np.reshape(a,(20*20)).shape
```

```python
np.reshape(b,(20*20,1)).shape
```

```python
a=np.zeros((20,20))
b=np.ones((20,20))
l2_dist(np.reshape(a,(20*20)), np.reshape(b,(20*20,1)))
```

```python
a=np.zeros((20,20))
b=np.ones((20,20))
l2_dist(np.reshape(a,(20*20)), np.reshape(b,(20*20)))
```

```python
a=np.zeros((20,20))
b=np.ones((20,20))
l2_dist(a.T,b.T)
```

```python
a=np.zeros((20,20))
b=np.ones((20,20))
l2_dist(a,b)
```
